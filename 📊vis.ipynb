# Import Module

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

print("Setup Compelite")
```

Load Dataset

```python
dataset = pd.read_csv('Status_Gizi_Bul.csv')
dataset.head(5)
```

VISUALISASI WITH SEABORN

```python
plt.figure(figsize=(10,4))
plt.title("Data Kendaraan Provinsi Bali")

sns.lineplot(data=dataset)
```

```python
dataset.columns
```

```python
import plotly.express as px
fig = px.bar(dataset, x="Kecamatan", y=dataset.columns,
              hover_data={"Kecamatan": "|%B %d, %Y"},
              title='VISUALISASI JUMLAH KENDARAAN DI PROVINSI BALI')
fig.update_xaxes(
    dtick="M1",
    tickformat="%b\n%Y")
fig.show()
```

```python
import plotly.graph_objects as go
kecamatan=dataset['Kecamatan']

fig = go.Figure(data=[
    go.Bar(name='Tinggi', x=kecamatan, y=[20, 14, 23]),
    go.Bar(name='Sedang', x=kecamatan, y=[12, 18, 29]),
    go.Bar(name='Pendek', x=kecamatan, y=[12, 18, 29])
])
# Change the bar mode
fig.update_layout(barmode='group')
fig.show()
```
